## 1. Selecting Series Data

```python
import numpy as np 
import pandas as pd
```

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.head(3)
```

```python
city = college['CITY']
city.head()
```

```python
#iloc selects location using integer.
city.iloc[3] #returns a string
```

```python
city.iloc[[0,1,2]] #returns a series
```

```python
#selecting a subset of values with the slice notation
city.iloc[4:50:10]
```

```python
#loc selections location using labels/strings.
city.loc['Alabama State University']
```

```python
np.random.seed(1)
labels = list(np.random.choice(city.index, 4))
labels
```

```python
#selection with loc indexer always includes the last element
city.loc['Alabama State University':'Reid State Technical College':10]
```

## 3. Selecting DF Rows and Columns simultaneously

```python
#this is the basic format for iloc and loc
"""
df.iloc[rows, columns]
here rows/cols can be (: (all)),(slice of rows/cols),(list of rows/cols - contiguous and otherwise), 
                      (variables containing list of integers for rows and cols)

df.loc[rows, columns]
here rows/cols can be the same options as iloc, but using labels/strings instead of integers.
"""
```

```python
#selecting a slice of rows and a single column
college.iloc[90:80:-2, 5]
#without specifying the step size, you would get an empty df as output
```

```python
#replicating above result with loc
#stop uses the value at 81, since loc would include the value at 80 as well(being "end inclusive")
start = college.iloc[90].name
stop = college.iloc[81].name
college.loc[start:stop:-2, 'RELAFFIL']
```

## 4. Selecting Data with both Integers and Labels

```python
#this could be done previously through ix indexer,
#but has since been deprecated due to the confusion surrounding it
col_start = college.columns.get_loc('UGDS_WHITE')
col_stop = college.columns.get_loc('UGDS_UNKN') + 1
col_start, col_stop

#get_loc takes an index label and returns its integer location
```

```python
college.iloc[:5, col_start:col_stop]
```

```python
#index[i] returns the label of the college at row i
row_start = college.index[10]
row_end = college.index[15]
college.loc[row_start:row_end, 'UGDS_WHITE':'UGDS_UNKN']
```

```python
#you COULD chain loc and iloc, but its generally a bad idea( i dont know why! )
college.iloc[10:16].loc[:, 'UGDS_WHITE':'UGDS_UNKN']
#this replicates the result obtained above
```

## 5. Speeding Up Scalar Selection

```python
#valuable if computation is time is of utmost importance
"""
improved speeds with .iat and .at in comparison to 
.iloc and .loc in the context of scalar selection
"""

#lets time loc and at
%timeit college.loc['Texas A & M University-College Station', 'UGDS_WHITE']
```

```python
#same result with at
%timeit college.at['Texas A & M University-College Station', 'UGDS_WHITE']
```

```python
row_num = college.index.get_loc('Texas A & M University-College Station')
col_num = college.columns.get_loc('UGDS_WHITE')
```

```python
%timeit college.iloc[row_num, col_num]
```

```python
%timeit college.iat[row_num, col_num]
```

## 7. Slicing Rows Lexicographically

```python
"""
loc lets you select all rows with an index lexicographically. 
Such a selection uses slice notation and works only if the index is sorted.
This is because the default behaviour of loc is select rows based on exact labels.
This behaviour is modified once the index col is sorted.
"""

college_sorted = college.sort_index()
college_sorted.head(3)
```

```python
#lets choose all colleges that lexicographically occur between Sp and Su,(aka, passing the slice)
#notice that we don't have to pass in exact label names now...
college_sorted.loc['Sp':'Su'].head()
```
